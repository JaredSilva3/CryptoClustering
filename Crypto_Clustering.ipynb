```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

---

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
scaler = StandardScaler() # Creates an instance of the StandardScaler 
df_market_data_scaled = scaler.fit_transform(df_market_data) # Scales the data to have a mean of 0 and a standard deviation of 1
```

```python
# Create a DataFrame with the scaled data
df_market_data_scaled = pd.DataFrame(df_market_data_scaled, columns=df_market_data.columns) # Converts the data into a DataFrame with the original column names 

# Copy the crypto names from the original DataFrame
df_market_data_scaled["coin_id"] = df_market_data.index # Adds 'coin_id' as a new column 

# Set the coin_id column as index
df_market_data_scaled = df_market_data_scaled.set_index("coin_id") # Makes 'coin_id' an index

# Display the scaled DataFrame
df_market_data_scaled.head() # Display the first 5 rows of the Dataframe
```

---

### Find the Best Value for k Using the Original Scaled DataFrame.

```python
# Create a list with the number of k-values from 1 to 11
k_values = list(range(1, 12)) # List of k values to test, from 1 to 11
```

```python
# Create an empty list to store the inertia values
inertia = [] # This will store the compactness of clusters for each k 

# Create a for loop to compute the inertia with each possible value of k
for k in k_values: # Loop through each k value  
    # 1. Create a KMeans model using the loop counter for the n_clusters
    model = KMeans(n_clusters=k, random_state=42) # Initialize the KMeans with k clusters 
    # 2. Fit the model to the data using `df_market_data_scaled`
    model.fit(df_market_data_scaled) # Fit the KMeans model to the scaled data 
    # 3. Append the model.inertia_ to the inertia list
    inertia.append(model.inertia_) # Add the inertia value to the list 
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data = {"k":k_values, "inertia":inertia} # Prepares data for elbow plot 

# Create a DataFrame with the data to plot the Elbow curve
df_elbow = pd.DataFrame(elbow_data) # Convert into a DataFrame 
```

```python
# Plot a line chart with all the inertia values computed with
# the different values of k to visually identify the optimal value for k.
df_elbow.hvplot.line ( # Use hvplot to create the Elbow curve
    x="k", # x-axis represents the number of clusters
    y="inertia", # y-axis represents inertia values
    title="Elbow Curve",
    xlabel="Number of Clusters (k)",
    ylabel="Inertia",
    width=800 # Width of the plot
)
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** 

The best value for `k` is **4** because the Elbow Curve flattens significantly at this point, indicating diminishing returns in reducing inertia with additional clusters

---

### Cluster Cryptocurrencies with K-means Using the Original Scaled DataFrame

```python
# Initialize the K-Means model using the best value for k
kmeans_model = KMeans(n_clusters=4, random_state=42) # Creates a KMeans model with k=4
```

```python
# Fit the K-Means model using the scaled DataFrame
kmeans_model.fit(df_market_data_scaled) # Fit KMeans model to the scaled data 
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled DataFrame
cluster_labels = kmeans_model.predict(df_market_data_scaled) # Predict cluster labels for each data point 

# Print the resulting array of cluster values.
print(cluster_labels)
```

```python
# Create a copy of the scaled DataFrame
df_clusters = df_market_data_scaled.copy()
```

```python
# Add a new column to the copy of the scaled DataFrame with the predicted clusters
df_clusters["Cluster"] = cluster_labels # Adds a new column called 'Cluster' with the cluster labels 

# Display the copy of the scaled DataFrame
df_clusters.head() # Display first 5 rows of the DataFrame with clusters 
```

```python
# Create a scatter plot using hvPlot by setting
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`.
# Color the graph points with the labels found using K-Means and
# add the crypto name in the `hover_cols` parameter to identify
# the cryptocurrency represented by each data point.
scatter_plot = df_clusters.hvplot.scatter(
    x="price_change_percentage_24h",
    y="price_change_percentage_7d",
    by="Cluster",
    hover_cols="coin_id",
    title="Cryptocurrency Clusters (k=4)"
)
```

---

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca = PCA(n_components=3) # PCA reduces the data to 3 principal components
```

```python
# Use the PCA model with `fit_transform` to reduce the original scaled DataFrame
# down to three principal components.
pca_data = pca.fit_transform(df_market_data_scaled) # Reduces the data dimensions to 3 components 

# View the scaled PCA data
print(pca_data[:5]) # Display the first 5 rows of the reduced PCA data 
```

```python
# Retrieve the explained variance to determine how much information
# can be attributed to each principal component.
explained_varience = pca.explained_variance_ratio_ # Retrieves the variance 
print(f"Explained Variance Ratios: {explained_varience}") # Prints the variance ratios 
print(f"Total Explained Variance: {explained_varience.sum()}") # Calculates the total variance 
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:** 

The total explained variance of the three principal components is **[`explained_variance.sum()`]**. This indicates the proportion of the original data's information retained after dimensional reduction.

```python
# Create a new DataFrame with the PCA data.
df_pca = pd.DataFrame(pca_data, columns=["PC1", "PC2", "PC3"]) # Create a new Dataframe with PCA components as columns 

# Copy the crypto names from the original scaled DataFrame
df_pca["coin_id"] = df_market_data_scaled.index # Adds the 'coin_id' from the original scaled DataFrame as a column

# Set the coin_id column as index
df_pca = df_pca.set_index("coin_id")

# Display the scaled PCA DataFrame
df_pca.head()
```

---

### Find the Best Value for k Using the Scaled PCA DataFrame

```python
# Create a list with the number of k-values from 1 to 11
k_values = list(range(1, 12)) # Generates a list of values from 1 to 11 for k 
```

```python
# Create an empty list to store the inertia values
inertia = [] # This list will store the inertia values for each k

# Create a for loop to compute the inertia with each possible value of k
for k in k_values: # Loop through each value of k in the list 
    
    # 1. Create a KMeans model using the loop counter for the n_clusters
    model = KMeans(n_clusters=k, random_state=42) # Initialize the KMeans model with k clusters 
    
    # 2. Fit the model to the data using `df_market_data_pca`
    model.fit(df_pca) # Fit the model using the transformed data 
    
    # 3. Append the model.inertia_ to the inertia list
    inertia.append(model.inertia_) # Save the inertia value to the list 
    
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data = {"k": k_values, "inertia": inertia} # Combine k-values and inertia into a dictionary 

# Create a DataFrame with the data to plot the Elbow curve
df_elbow = pd.DataFrame(elbow_data) # Converts dictionary into a DataFrame 

# Display the Dataframe to verify 
df_elbow.head() # Displays the first 5 row of the Elbow DataFrame
```

```python
# Plot a line chart with all the inertia values computed with
# the different values of k to visually identify the optimal value for k.
df_elbow.hvplot.line( # Use hvPlot to create a line chart
    x="k", # x-axis: number of cluster (k-values)
    y="inertia", # y-axis: inertia values
    title="Elbow Curve",
    xlabel="Number of Cluster (k)",
    ylabel="Inertia",
    width=800
)
```

#### Answer the following questions: 

* **Question:** What is the best value for `k` when using the PCA data?

  * **Answer:**

  The best value for `k` is **4** because the Elbow Curve flattens significantly at this point, indicating minimal improvement in inertia with additional clusters


* **Question:** Does it differ from the best k value found using the original data?

  * **Answer:** 

  No, it does not differ. The best value for `k` remains **4** for both the original scaled data and the PCA data.

### Cluster Cryptocurrencies with K-means Using the Scaled PCA DataFrame

```python
# Initialize the K-Means model using the best value for k
kmeans_pca = KMeans(n_clusters=4, random_state=42) # Create a KMeans model with k=4
```

```python
# Fit the K-Means model using the PCA data
kmeans_pca.fit(df_pca) # Train the KMeans model using the transformed data 
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled PCA DataFrame
pca_clusters = kmeans_pca.predict(df_pca) # Predict cluster labels for each cryptocurrency 

# Print the resulting array of cluster values.
print(pca_clusters)
```

```python
# Create a copy of the scaled PCA DataFrame
df_pca_clusters = df_pca.copy() # Make a copy of the PCA DataFrame to avoid changing the original data

# Add a new column to the copy of the PCA DataFrame with the predicted clusters
df_pca_clusters["Cluster"] = pca_clusters # Add a 'Cluster' label as a new column 

# Display the copy of the scaled PCA DataFrame
df_pca_clusters.head()
```

```python
# Create a scatter plot using hvPlot 
scatter_plot_pca = df_pca_clusters.hvplot.scatter(  # Create a scatter plot using hvPlot
    x="PC1",  # x-axis: Principal Component 1
    y="PC2",  # y-axis: Principal Component 2
    by="Cluster",  # Color the points based on their cluster labels
    hover_cols=["coin_id"],  # Display the crypto name (coin_id) when hovering over points
    title="Cryptocurrency Clusters (PCA Data)"  # Title of the scatter plot
)

# Display the scatter plot
scatter_plot_pca
```
